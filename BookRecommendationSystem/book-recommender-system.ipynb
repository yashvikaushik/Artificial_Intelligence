POPULARITY BASED RECOMMENDER SYSTEM


```python
ratings_with_name=ratings.merge(books,on='ISBN')
ratings_with_name.head()
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_Rating'},inplace=True)
num_rating_df

```

```python
avg_rating_df = (ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index())
avg_rating_df.rename(columns={'Book-Rating':'avg_Rating'},inplace=True)
avg_rating_df.head()
```

```python
popularity_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popularity_df.head()
```

```python
popularity_df=popularity_df[popularity_df['num_Rating']>=250]
popularity_df.head()
```

```python
popularity_df.sort_values(['avg_Rating'],ascending=False)
```

```python
popularity_df=popularity_df.sort_values(['avg_Rating'],ascending=False).head(100)
```

COLABORATIVE FILTERING BASED RECOMMENDER SYSYTEM

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
```

```python
inteligent_users=x[x].index
print(inteligent_users)
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(inteligent_users)]
filtered_rating.head()
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
y[y] #boolean indexing
```

```python
famous_books=y[y].index
famous_books
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_rating.drop_duplicates()
```

```python
pivotTable=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivotTable.fillna(0,inplace=True)
```

```python
pivotTable
```

now this table contains users which have rated more than 200 books and contains the books which has minimum 50 ratings


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pivotTable)
```

```python
similarity_scores[0]
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    #index fetch 
    index=np.where(pivotTable.index==book_name)[0][0]
    #similar items 
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:11]
    for i in similar_items:
        print(pivotTable.index[i[0]])
    return 
```
