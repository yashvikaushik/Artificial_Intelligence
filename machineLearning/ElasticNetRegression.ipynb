```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import ElasticNet,LinearRegression,Ridge,Lasso 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import r2_score
```

```python
X,y=load_diabetes(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.2,random_state=2)
```

```python
#LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
#Ridge regularization 
r=Ridge(alpha=0.1)
r.fit(X_train,y_train)
y_predr=r.predict(X_test)
r2_score(y_test,y_predr)
```

```python
#Lasso 
l=Lasso(alpha=0.01)
l.fit(X_train,y_train)
predl=l.predict(X_test)
r2_score(y_test,predl)
```

```python
#elastic net 
e=ElasticNet(alpha=0.005,l1_ratio=0.9)
e.fit(X_train,y_train)
prede=e.predict(X_test)
r2_score(y_test,prede)
```

“Elastic Net achieved the highest R² score, indicating that a combination of L1 and L2 regularization is most suitable for this dataset. Ridge regression also improved performance over linear regression, suggesting multicollinearity among features. Lasso performed slightly worse due to its aggressive feature selection.”
