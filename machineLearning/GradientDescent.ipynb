```python
from sklearn.datasets import make_regression 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import cross_val_score 
```

```python
X,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=20,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.intercept_
```

```python
y_pred=lr.predict(X_test)
from sklearn.metrics import r2_score 
r2_score(y_test,y_pred)
```

```python
class GDRegressor:
    def __init__(self,learning_rate,epoch):
        self.m=100
        self.b=-120
        self.learning_rate=learning_rate
        self.epoch=epoch

    def fit(self,X,y):
        #calculate b using gradient descent
        for i in range(self.epoch):
            loss_b_slope=-2*np.sum((y-self.m*X.ravel()-self.b)) #slope of loss function for b 
            loss_m_slope=-2*np.sum((y-self.m*X.ravel()-self.b)*X.ravel())
            
            self.b=self.b-(self.learning_rate*loss_b_slope)
          
        print(self.b,self.m)

    def predict(self,X):
        return self.m*X + self.b
     
```

```python
gd=GDRegressor(0.01,10)
```

```python
gd.fit(X_train,y_train) #graph is stuck in a zig zag loop condidtion a learning rate issue
```
